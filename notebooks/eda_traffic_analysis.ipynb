```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
file_path = "/Users/sohamsolanki/Desktop/Calgary_Traffic_Analysis/data/Cleaned_Traffic_data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to confirm it's loaded
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define save paths
save_folder = "/Users/sohamsolanki/Desktop/Calgary_Traffic_Analysis/dashboard"

# Incidents Per Year
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Year'], hue=df['Year'], palette="Blues", legend=False)
plt.title("Number of Traffic Incidents Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.savefig(save_folder + "incidents_per_year.png")
plt.show()

# Incidents Per Month
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Month'], hue=df['Month'], palette="coolwarm", legend=False)
plt.title("Traffic Incidents by Month")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(range(1, 13))
plt.savefig(save_folder + "incidents_by_month.png")
plt.show()

# Incidents by Day of the Week
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Weekday'], hue=df['Weekday'], 
              order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
              palette="Set2", legend=False)
plt.title("Traffic Incidents by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.savefig(save_folder + "incidents_by_day.png")
plt.show()

# Peak Traffic Incident Hours
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Hour'], hue=df['Hour'], palette="magma", legend=False)
plt.title("Traffic Incidents by Hour of the Day")
plt.xlabel("Hour (24-hour format)")
plt.ylabel("Count")
plt.savefig(save_folder + "incidents_by_hour.png")
plt.show()
```
